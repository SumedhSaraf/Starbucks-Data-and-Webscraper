```python
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
import pandas as pd
import urllib.request
import datetime

pd.set_option('display.max_rows', 2500)
resp = urllib.request.urlopen('https://www.apple.com/')
soup = BeautifulSoup(resp, from_encoding=resp.info().get_param('charset'))
l1 = []
l1_d = []
for link in soup.find_all('a', href=True):
    if (link['href'].startswith('https') or link['href'].startswith('www')):
        l1.append(link['href'])
        l1_d.append(link.text)
      
        
    if link['href'].startswith('//www'):
        l1.append(link['href'][2:])
        l1_d.append(link.text)
    
        
l2 = []
l2_d = []
for links in soup.find_all('link', href=True):
    if (links['href'].startswith('https') or links['href'].startswith('www')):
        l2.append(links['href'])
        if (links.has_attr('hreflang')):
            l2_d.append(links['hreflang'])
        else:
            l2_d.append("unavailable")

        
    if links['href'].startswith('//www'):
        l2.append(links['href'])
        if (links.has_attr('hreflang')):
            l2_d.append(links['hreflang']) 
        else:   
            l2_d.append("unavailable")   
            
l3 = l1 + l2
l3_d = l1_d + l2_d          

def get_status(url):
    http = urllib3.PoolManager()
    response = http.request('GET',url)
    if response.status == 200 or response.status < 300:
         x = 1
    else:
         x = 0
    return x    
    
df = pd.DataFrame(columns = ['link','description','validity','Checkingtime'])
for y in range(0,len(l3)-1):
    df2 = pd.DataFrame([[l3[y],l3_d[y],get_status(l3[y]),str(datetime.datetime.now())]], columns= ['link','description','validity','Checkingtime'])
    df = df.append(df2) 
df.to_csv('output1.csv',date_format='%Y-%m-%d %H:%M:%S')    
```
